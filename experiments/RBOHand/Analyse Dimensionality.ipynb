```julia
path = "/Users/zahedi/projects/TU.Berlin/dims/"
dirs = readdir(path);
dirs = filter(x->contains(x,"hand"),dirs)
dirs = filter(x->contains(x, "_"),dirs);
```

```julia
control_dims_files         = map(x->"$path/$x/control.states.csv.dim",dirs)
control_dims               = map(x->readcsv(x),control_dims_files)
control_dim_MLE            = map(x->x[1,2], control_dims)
control_dim_CorrDim        = map(x->x[2,2], control_dims)
control_dim_NearNb         = map(x->x[3,2], control_dims)
control_dim_PackingNumbers = map(x->x[4,2], control_dims)
control_dim_GMST           = map(x->x[5,2], control_dims)
control_dim = DataFrame()
control_dim[:Names] = map(x->split(x,"/")[8],control_dims_files)
control_dim[:MLE] = control_dim_MLE
control_dim[:CorrDim] = control_dim_CorrDim
control_dim[:NearNb] = control_dim_NearNb
control_dim[:PackingNumbers] = control_dim_PackingNumbers
control_dim[:GMST] = control_dim_GMST
writetable("$path/control.states.tabel", control_dim)
control_dim
```

```julia
hand_sofastates_dims_files         = map(x->"$path/$x/normalised.hand.sofastates.csv.dim",dirs)
hand_sofastates_dims               = map(x->readcsv(x),hand_sofastates_dims_files)
hand_sofastates_dim_MLE            = map(x->x[1,2], hand_sofastates_dims)
hand_sofastates_dim_CorrDim        = map(x->x[2,2], hand_sofastates_dims)
hand_sofastates_dim_NearNb         = map(x->x[3,2], hand_sofastates_dims)
hand_sofastates_dim_PackingNumbers = map(x->x[4,2], hand_sofastates_dims)
hand_sofastates_dim_GMST           = map(x->x[5,2], hand_sofastates_dims)
hand_sofastates = DataFrame()
hand_sofastates[:Names] = map(x->split(x,"/")[8], hand_sofastates_dims_files)
hand_sofastates[:MLE] = hand_sofastates_dim_MLE
hand_sofastates[:CorrDim] = hand_sofastates_dim_CorrDim
hand_sofastates[:NearNb] = hand_sofastates_dim_NearNb
hand_sofastates[:PackingNumbers] = hand_sofastates_dim_PackingNumbers
hand_sofastates[:GMST] = hand_sofastates_dim_GMST
writetable("$path/hand.sofastates.tabel", hand_sofastates)
writetable("/Users/zahedi/Desktop/intrinsic.dimension.csv", hand_sofastates)
hand_sofastates
```

```julia
states                     = DataFrame()
states[:Names]             = map(x->split(x,"/")[8], hand_sofastates_dims_files)
states[:MLE_HS]            = hand_sofastates[:MLE]
states[:MLE_CS]            = control_dim[:MLE]

states[:CorrDim_HS]        = hand_sofastates[:CorrDim]
states[:CorrDim_CS]        = control_dim[:CorrDim]

states[:NearNb_HS]         = hand_sofastates[:NearNb]
states[:NearNb_CS]         = control_dim[:NearNb]

states[:PackingNumbers_HS] = hand_sofastates[:PackingNumbers]
states[:PackingNumbers_CS] = control_dim[:PackingNumbers]

states[:GMST_HS]           = hand_sofastates[:GMST]
states[:GMST_CS]           = control_dim[:GMST]

writetable("$path/experiment.values.csv", states)
states
```

```julia
dirs = readdir(path);
dirs = filter(x->contains(x, "hand"),dirs);
dirs = filter(x->contains(x, "_"),dirs);
pattern = unique(map(x->replace(x, r"(object\w+_).+",s"\g<1>"), dirs))

control_dim_MLE_mean = []
control_dim_MLE_std  = []

control_dim_CorrDim_mean = []
control_dim_CorrDim_std  = []

control_dim_NearNb_mean    = []
control_dim_NearNb_std     = []
    
control_dim_PackingNumbers_mean = []
control_dim_PackingNumbers_std  = []

control_dim_GMST_mean   = []
control_dim_GMST_std    = []


for p in pattern
    ps = filter(x->contains(x,p), dirs)
    control_dims               = map(x->readcsv("$path/$x/control.states.csv.dim"), ps)
    control_dim_MLE            = map(x->x[1,2], control_dims)
    control_dim_MLE_mean       = vcat(control_dim_MLE_mean, mean(control_dim_MLE))
    control_dim_MLE_std        = vcat(control_dim_MLE_std,  std(control_dim_MLE))

    control_dim_CorrDim        = map(x->x[2,2], control_dims)
    control_dim_CorrDim_mean   = vcat(control_dim_CorrDim_mean, mean(control_dim_CorrDim))
    control_dim_CorrDim_std    = vcat(control_dim_CorrDim_std,  std(control_dim_CorrDim))

    control_dim_NearNb         = map(x->x[3,2], control_dims)
    control_dim_NearNb_mean    = vcat(control_dim_NearNb_mean, mean(control_dim_NearNb))
    control_dim_NearNb_std     = vcat(control_dim_NearNb_std,  std(control_dim_NearNb))
    
    control_dim_PackingNumbers      = map(x->x[4,2], control_dims)
    control_dim_PackingNumbers_mean = vcat(control_dim_PackingNumbers_mean, mean(control_dim_PackingNumbers))
    control_dim_PackingNumbers_std  = vcat(control_dim_PackingNumbers_std,  std(control_dim_PackingNumbers))
    
    control_dim_GMST        = map(x->x[5,2], control_dims)
    control_dim_GMST_mean   = vcat(control_dim_GMST_mean, mean(control_dim_GMST))
    control_dim_GMST_std    = vcat(control_dim_GMST_std,  std(control_dim_GMST))


end

control_states                = DataFrame()
control_states[:Names]        = map(x->replace(x,r"_$",s""),pattern)

control_states[:MLE_mean]     = round(control_dim_MLE_mean,3)
control_states[:MLE_std]      = round(control_dim_MLE_std,3)

control_states[:CorrDim_mean] = round(control_dim_CorrDim_mean,3)
control_states[:CorrDim_std]  = round(control_dim_CorrDim_std,3)

control_states[:NearNb_mean]  = round(control_dim_NearNb_mean,3)
control_states[:NearNb_std]   = round(control_dim_NearNb_std,3)

control_states[:PackingNumbers_mean]  = round(control_dim_PackingNumbers_mean,3)
control_states[:PackingNumbers_std]   = round(control_dim_PackingNumbers_std,3)

control_states[:GMST_mean]    = round(control_dim_GMST_mean,3)
control_states[:GMST_std]     = round(control_dim_GMST_std,3)

writetable("$path/control.states.table.averages", control_states)
control_states
```
